```python
# -----------------------
# Example usage (driver)
# -----------------------

if __name__ == "__main__":

    from Asymmetric_Generalization import (
        read_data_stat, prefilter_orderbook_features, orderbook_split_stat,
        compute_regression_metrics, compute_quantile_crossing_rate_regressor, get_features,
        train_status
    )

    save_path = train_status('local')         
    indice     = 'ID3'
    quantiles  = [0.1, 0.5, 0.9]
    selection  = 'top_5'                        
    resolutions = ['h', 'qh']
    markets     = ['germany', 'austria']

    mlp_cfg = dict(
        hidden=(128, 64, 32),
        dropout=0.10,
        lr=1e-3,
        epochs=50,
        batch_size=1024,
        seed=42,
    )

    df_country = run_cross_country_experiments(
        save_path, indice, selection, quantiles, resolutions,
        read_data_stat, prefilter_orderbook_features, orderbook_split_stat,
        compute_regression_metrics, compute_quantile_crossing_rate_regressor, get_features,
        mlp_cfg
    )
    print("Cross-country results saved:", os.path.join(save_path, "Result", "cross_country_results_qMLP.csv"))
    print(df_country)

    df_product = run_cross_product_type_experiments(
        save_path, indice, selection, quantiles, markets,
        read_data_stat, prefilter_orderbook_features, orderbook_split_stat,
        compute_regression_metrics, compute_quantile_crossing_rate_regressor, get_features,
        mlp_cfg
    )
    print("Cross-product-type results saved:", os.path.join(save_path, "Result", "cross_product_type_results_qMLP.csv"))
    print(df_product)
```
